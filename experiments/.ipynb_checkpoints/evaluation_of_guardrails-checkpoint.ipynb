```python
import evaluate
import pandas as pd
eval_df = pd.read_csv("evalset_with_response.csv")
eval_df
```

```python
## bleu scoring

bleu = evaluate.load("bleu")
def compute_bleu(answer, response):
    results = bleu.compute(predictions = [response], references = [answer])
    return results['bleu']
eval_df['bleu_score'] = eval_df.apply(lambda x: compute_bleu(x['response'], x['Answer']), axis = 1)
eval_df
```
